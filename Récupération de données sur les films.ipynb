Explication : 

Un client va depuis une application ou un navigateur solliciter les serveurs d'une entreprise pour répondre à une demande de service. Pour cela il va utiliser le protocole HTTP (Hypertext Transfert Protocole) et cibler une URL. En fonction de l'action (GET, POST, DELETE, PUT), les serveurs vont lui donner autorisation ou non. En conséquence, un fichier JSON va être généré, donnant la réponse.

L'API est donc la "passerelle" entre la demande cliente et la réponse serveur

# Le JSON

Définissez ce qu’est JSON, comment manipuler avec python une réponse JSON ? Définissez paramètres d'une requête, découvrez comment placer ces paramètres soit dans une url, soit avec la méthode get() de requests.

JSON, ou JavaScript Object Notation, est un format de données léger et facile à lire utilisé pour l'échange de données entre applications. Il est basé sur un sous-ensemble du langage JavaScript, mais il est indépendant de la plate-forme, ce qui signifie qu'il peut être utilisé avec divers langages de programmation. JSON est couramment utilisé pour représenter des données structurées, telles que des objets et des tableaux, et est souvent utilisé dans les services web, les API et la communication entre serveurs et clients.

```python
# En Python, vous pouvez manipuler une réponse JSON de plusieurs manières, notamment en utilisant le module json intégré à Python. Voici comment faire :

# Importez le module json :

import json

# Chargez une réponse JSON : Si vous avez une réponse JSON sous forme de chaîne de caractères (par exemple, reçue en réponse à une requête HTTP), vous pouvez la convertir en un objet Python en utilisant la méthode json.loads() (load string) :

json_string = '{"name": "John", "age": 30, "city": "New York"}'
data = json.loads(json_string)

# Accédez aux données JSON : Vous pouvez maintenant accéder aux données JSON comme si elles étaient des dictionnaires Python :

print(data["name"])  # Affiche "John"
print(data["age"])   # Affiche 30
print(data["city"])  # Affiche "New York"

# Convertissez un objet Python en JSON : Vous pouvez également convertir un dictionnaire Python en une chaîne JSON à l'aide de la méthode json.dumps() (dump string) :

data = {"name": "Alice", "age": 25, "city": "Los Angeles"}
json_string = json.dumps(data)
```

```python
# En ce qui concerne la définition des paramètres d'une requête HTTP, vous pouvez les transmettre à un serveur de deux manières courantes :

# Dans l'URL : Vous pouvez ajouter les paramètres à l'URL de la requête en les concaténant à l'URL sous la forme de chaînes de requête. Par exemple :

# https://example.com/api/resource?param1=value1&param2=value2

# En Python, vous pouvez construire une URL avec des paramètres de requête en utilisant la bibliothèque urllib.parse :
from urllib.parse import urlencode

params = {
    "param1": "value1",
    "param2": "value2"
}

query_string = urlencode(params)
url = "https://example.com/api/resource?" + query_string
# Avec la méthode get() de la bibliothèque requests : Vous pouvez également inclure les paramètres de la requête en utilisant la méthode get() de la bibliothèque requests en tant que dictionnaire. Par exemple :

import requests

params = {
    "param1": "value1",
    "param2": "value2"
}

response = requests.get("https://example.com/api/resource", params=params)

# La bibliothèque requests se charge de construire correctement l'URL avec les paramètres de requête pour vous.
```
